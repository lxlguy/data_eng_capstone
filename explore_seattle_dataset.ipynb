```python
import json

with open('credentials.json') as f:
    data = json.load(f)
    socrates_id = data['socrates_id']
    socrates_key = data['socrates_api_key']
```

```python
import pandas as pd
from sodapy import Socrata
from IPython.core.display import display, HTML

display(HTML("<style>.container { width:100% !important; }</style>"))
pd.set_option('display.max_colwidth', None)
```

```python
inventory_db = "6vkj-f5xf"
loans_db = "5src-czff"

client = Socrata("data.seattle.gov", None, username=socrates_id,password=socrates_key, timeout=50)

# Example authenticated client (needed for non-public datasets):
# client = Socrata(data.seattle.gov,
#                  MyAppToken,
#                  userame="user@example.com",
#                  password="AFakePassword")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get(inventory_db, limit=5000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```

```python
results = client.get(loans_db, limit=5000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```

```python
results = client.get(inventory_db, where='itemlocation="hip"', limit=100000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```

```python
len(results_df)
```

```python
results = client.get(inventory_db, where='itemlocation="hip"', limit=100)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```

```python
results = client.get(inventory_db, where='itemlocation="hip"', limit=100, offset=1)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```

```python
meta = client.get_metadata(inventory_db)
meta
```

```python
results = client.get(inventory_db, select="itemlocation, count(*)", group = "itemlocation")

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
results_df.head()
```
